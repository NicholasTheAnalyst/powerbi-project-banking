```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Medium', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df.head()
```

```python
# Визуализация распределения клиентов по уровням доходов
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Рассмотрю распределение уникальных значений в категориальных столбцах
```

```python
categorical_cols = df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}:'")
    display(df[col].value_counts())
```

```python
# Одномерный анализ
```

```python
for col in categorical_cols:
    plt.figure()  
    sns.countplot(data=df, x=col)
    plt.show()
```

```python
# Двумерный анализ
```

```python
for col in categorical_cols:
    plt.figure()  
    sns.countplot(data=df, x=col, hue='GenderId')
```

```python
for col in categorical_cols:
    plt.figure()  
    sns.countplot(data=df, x=col, hue='Nationality')
```

```python
# Автоматическое построение отсортированных диаграмм распределения для категориальных признаков
```

```python
for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(9, 5))
    sns.countplot(
        data=df, 
        x=col, 
        order=df[col].value_counts().index
    )
    plt.title(f'Distribution of {col}', fontsize=14, pad=15)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count (Frequency)', fontsize=12)

    plt.show()
```

```python
# Распределение числовых признаков (доход и накопления)
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings']
```

```python
plt.figure(figsize=(15,10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 1, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.show()
```

```python
# Построение тепловой карты корреляции для числовых признаков
```

```python
numerical_cols_all = ['Estimated Income', 'Superannuation Savings', 'Age', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account', 'Business Lending']
```

```python
correlation_matrix = df[numerical_cols_all].corr()

plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
